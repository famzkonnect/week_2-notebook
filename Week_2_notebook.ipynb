```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
data = pd.read_csv("C:/Users/MY LAPTOP/Documents/Hamoye Internship/WEEK 2/energydata_complete.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
# Drop date and lights
data.drop(["date", "lights"], axis=1, inplace=True)
```

```python
data.head()
```

```python
#normalize the dataset
scaler = MinMaxScaler()
data = pd.DataFrame(scaler.fit_transform(data),  columns = data.columns)
```

```python
#get features and labels
x = data.drop(columns=['Appliances'])

y = data['Appliances']
```

```python
# Separate train and test dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
#select a sample of the dataset
reg_df = data[['T2', 'T6']]

reg_df.head()
```

```python
#reshape sample dataset
x= reg_df['T2'].values.reshape(-1,1)
y = reg_df['T6'].values.reshape(-1,1)
```

```python
#split sample dataset into train and test sets
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
linear_regr = linear_model.LinearRegression() 
```

```python
# Train the model using the training sets
linear_regr.fit(xtrain, ytrain)
```

```python
# Make predictions using the testing set
pred = linear_regr.predict(xtest)
```

```python
#R-squared or Coefficient of determination
r2_score = r2_score(ytest, pred)
print('R-squared:',(round(r2_score, 2)))
```

```python
mae = mean_absolute_error(y_test, pred)

print('MAE:',(round(mae, 3)))
```

```python
rmse = np.sqrt(mean_squared_error(y_test, ytrain))
print('RMSE:',(round(rmse, 3)))
```
